## Måle makroskopiske størrelser

### Temperatur

Vi kan måle temperaturen ved å se på ekvipartisjonsprinsippet:

\begin{equation}
\frac{3}{2}Nk_BT = \langle K \rangle
\end{equation}

Vi kan da finne temperaturen ved

\begin{equation}
T = \frac{2\langle K\rangle}{3Nk_B}
\end{equation}

Vi finner gjennomsnittet av kinetisk energi ved

\begin{equation}
    \langle K \rangle = \langle \sum_i \frac{1}{2}m_i v_i^2 \rangle.
\end{equation}

Lammps bruker samme metode.

### Trykk

Vi kan måle gjennomsnittlig trykk

\begin{equation}
    P = \rho k_B T + \frac{1}{3V} \sum_{i<j}\mathbf{F}_{ij}\cdot \mathbf{r}_{ij}
\end{equation}

hvor summen går over partikkel par som interagerer. 

I Lammps finner vi trykket med compute pressure, som bruker

\begin{equation}
P = \frac{Nk_B T}{V} + \sum_i^{N}\frac{\mathbf{r}_i \cdot f_i}{dV}
\end{equation}

hvor andre ledd er virialet. Summen går over antall partikler.

### Mean squared displacement

Vi kan regne ut msd ved

\begin{equation}
    \langle r^2(t) \rangle  = \frac{1}{N} \sum_{i=1}^N (\mathbf{r}(t) - \mathbf{r}_0)^2
\end{equation}

I Lammps finner vi msd ved

\begin{equation}
\langle r^2 \rangle = dx^2 + dy^2 + dz^2
\end{equation}

### Utfordringer

Hvis vi bruker en termostat som skalerer hastighetene samtidig som du samler inn data for å regne ut msd vil man kunne få feil diffusjonskonstant, ettersom skaleringen av hastighetene kan føre til ufysisk oppførsel.

### Bruksområder

msd kan brukes til å finne diffusjonskonstanten. Vi har at 

\begin{equation}
    \langle r^2 \rangle = (2d)Dt \quad t \rightarrow \infty
\end{equation}

hvor d er dimensjonen ($2d = 6$ for 3d). Ved å gjøre en lineær tilpasning på msd kan vi finne stigningstallet $D' = D/6$ til $\langle r^2(t) \rangle$. Når vi har funnet diffusjonkonstanten som funksjon av temperatur kan vi observere faseoverganger, som vist i figuren under for Si og vann. Vi ser også diffusjonskonstanten for Argon.
