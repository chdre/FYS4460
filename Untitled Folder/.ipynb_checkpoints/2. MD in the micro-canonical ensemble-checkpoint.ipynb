# MD in the micro-canonical ensemble

### Initialisering

Vi initialiserer systemet ved å velge posisjoner og hastigheter til partiklene. For potensialer som eksempelvis LJ er det viktig at vi ikke plasserer partiklene for nærme, ettersom dette vil resultere i store frastøtningskrefter, som igjen vil være til at hastighetene blir veldig store når vi begynner å integrere systemet. 

Vi plasserer atomene jevnt i systemet ved å kontruere en unit cell med partikler. Unit cell blir så kopiert slik at den fyller hele systemet. 

For LJ vet vi at krystallstrukturen er lik en face centered cube. En FCC er slik at vi har en partikkel i $(0,0,0)$ og en partikkel i midten av flatene tilknyttet $(0,0,0)$.

### Effekter av initialisering

Når vi initialiserer systemet med fastsatte posisjoner og uniformt eller Gaussisk fordelte hastighete er systemet ufysisk. Vi må derfor la systemet gå i likevekt før vi begynner å gjøre utregninger, eksempelvis ved å termalisere eller integrere en gitt tid. Derfor vi ikke velger temperatur og tetthet med omhu kan vi risikere at det tar lengre tid før systemet kommer i likevekt. 

### Temperaturmål og fluktuasjoner

Vi kan måle temperaturen ved å se på ekvipartisjonsprinsippet:

\begin{equation}
\frac{3}{2}Nk_BT = \langle K \rangle
\end{equation}

Vi kan da finne temperaturen ved

\begin{equation}
T = \frac{2\langle K\rangle}{3Nk_B}
\end{equation}

Vi finner gjennomsnittet av kinetisk energi ved

\begin{equation}
    \langle K \rangle = \langle \sum_i \frac{1}{2}m_i v_i^2 \rangle.
\end{equation}

Ettersom temperaturen er proporsjonal med den snittet av den kinetiske energien, som igjen er proporsjonal med snittet hastighetene forventer vi at den fluktuerer. 

Dersom vi har et grovt tidssteg vil vi risikere at to partikler kommer veldig nær hverandre, som vil bety at partiklene får høye hastigheter. Vi vil da forvente at det tar lenger tid for systemet å komme i likevekt og at temperaturen vil fluktuere veldig.

### Termostater

En termostat har som hensikt å hjelpe systemet å få en gitt temperatur. Det kan benyttes for å stabilisere temperaturen eller å heve/senke den.

Vi ser for oss at vi kobler et varmebad til systemet og spesifiserer temperaturen dens. Målet til termostaten er da å holde systemtemperaturen på samme temperatur som varmebadet.

Noen termostater skalerer hastighetene til partiklene i systemet. Når hastighetene blir skalert er det et brudd med det kanoniske system. Nosé-Hoover erstatter varmebadet med en imaginær partikkel og den produserer resultater som er realistiske og i tråd med det kanoniske ensemble. Istedenfor å skalere hastighetene endrer Nosé-Hoover bevegelseslikningene.

Nedenfor kan vi se en sammenlikning av temperaturfluktuasjoner for Berendsen og Nose-Hoover for et Argon-system (oppg. fra prosjekt 1).
