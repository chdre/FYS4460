### Discuss the percolation problem on a 2 Ã— 2 lattice

```python
import numpy as np
import scipy.ndimage as scpi
import matplotlib.pyplot as plt

np.random.seed(42)
p = 0.6
L = 2
domain = np.random.rand(L, L)
binary_domain = domain < p
label, num = scpi.measurements.label(binary_domain)
```

```python
plt.imshow(binary_domain, origin='lower')
plt.title(f'p = {p}, L={L}')
plt.show()
```

Dersom vi ser pÃ¥ perkolasjon pÃ¥ en 2x2 lattice er det $2^4=16$ forskjellige konfigurasjoner av sites som er okkupert eller ikke.


I figuren ser vi de forskjellige kombinasjonene av okkupert (sort) og ikke-okkupert (hvit). Grunnet symmetri og det at systemet kan perkolere i bÃ¥de x og y har flere av konfigurasjonene lik sannsynlighet for perkolasjon og for at et site tilhÃ¸rer en spanning cluster, vi kan derfor gruppere disse konfigurasjonene, men vi mÃ¥ da huske multiplisiteten $g_c$. Vi ser at for $c=3, 5, 6$ er perkolasjonssannsynligheten for p og L gitt konfigurasjonen c; $\Pi(p,L|c) = 1$, og for de Ã¸vrige konfigurasjonene er den 0. 

Vi har sannsynligheten $p$ for at et site er okkupert. Sannsynligheten for at et site er ikke-okkupert er da $1-p$. 

```python
p = np.linspace(0, 1, 100)
Pi = 4 * p ** 2 * (1 - p) ** 2 + 4 * p ** 3 * (1 - p) + p ** 4

plt.plot(p, Pi)
plt.show()
```
