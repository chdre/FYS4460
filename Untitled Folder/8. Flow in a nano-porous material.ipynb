### Diskuter hvordan man induserer flow i et nano-porøst materiale:
Vi induserer en strømning i et nanoporøst materiale ved å introdusere en ekstrern kraft som virker på hver partikkel i fluidet. For en strømning i et gravitasjonsfelt har vi fra Darcy's lov $U = \frac{k}{\mu}(\nabla P - \rho g)$, hvor $\rho$ er massetettheten til fluidet og $g$ er tyngdeaksellerasjonen.

Vi kan erstatte $\rho g$ med den eksterne kraften som virker på hver partikkel: 
\begin{equation}
    \rho g = \frac{M}{V}g = \frac{N}{V}mg = \frac{N}{V}G = \frac{N}{V}F_x = nF_x.
\end{equation}

Dette gir at strømningsprofilen blir $U = \frac{k}{\mu}(\nabla P - nF_x)$


### Hvordan kan du sjekke modellen:
Vi vet at for en sylindrisk pore i et ellers fast materiale forventer vi at en strøm har en parabolisk form tilsvarende:

\begin{equation}
u(r) = \frac{G}{4\mu}(a^2 - r^2) = \frac{F_x}{4\mu}(a^2 - r^2)
\end{equation}

hvor vi har satt inn $G = F_x$. 



Her kan vi se flow profilen, hvor teoretisk flow

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 1000)
r = 1
G = 1
mu = 1

plt.plot(r*r - x*x, x)
plt.title(f'Flow profile')
plt.legend(['$v(x)$'])
plt.xlabel('x')
plt.ylabel('$r^2 - x^2$')
plt.show()
```
