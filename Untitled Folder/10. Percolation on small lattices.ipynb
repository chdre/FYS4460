### Discuss the percolation problem on a 2 Ã— 2 lattice

```python
import numpy as np
import scipy.ndimage as scpi
import matplotlib.pyplot as plt

np.random.seed(42)
p = 0.6
L = 2
domain = np.random.rand(L, L)
binary_domain = domain < p
label, num = scpi.measurements.label(binary_domain)
```

Vi ser at for $c=3, 5, 6$ er perkolasjonssannsynligheten for p og L gitt konfigurasjonen c; $\Pi(p,L|c) = 1$, og for de Ã¸vrige konfigurasjonene er den 0.

Tegn figur som viser at nÃ¥r L Ã¸ker, sÃ¥ blir $\Pi$ nÃ¦rmere og nÃ¦rmere en step-funksjon om $p_c$. NÃ¥r L Ã¸ker sÃ¥ gÃ¥r $P(p, L)$ mot en funksjon som er null frem til $p_c$, Ã¸ker sÃ¥ ( til lineÃ¦r oppfÃ¸rsel ved $\approx 0.7$.

```python
L = np.array([2, 4, 8, 16, 32, 64, 128])
p = np.linspace(0, 1, 100)
N = 1000


def P_func(p, L, N):
    nx = len(p)
    nL = len(L)

    P = np.zeros((nx, nL))
    
    for k, l in enumerate(L):
        for i in range(N):
            domain = np.random.rand(l, l)
            for j, pi in enumerate(p):
                binary_domain = domain < pi
                label, num = scpi.measurements.label(binary_domain)
                perc_x = np.intersect1d(label[0, :], label[-1, :])
                perc = perc_x[perc_x > 0]
                if len(perc) > 0:
                    area = scpi.measurements.sum(binary_domain, label, perc[0])
                    P[j, k] += area

        P[:, k] /= (N * l * l)

    return P

        
def Pi_func(p, L, N):
    nx = len(p)
    nL = len(L)

    Ni = np.zeros((nx, nL))
    Pi = np.zeros((nx, nL))
    
    for k, l in enumerate(L):
        for i in range(N):
            domain = np.random.rand(l, l)
            for j, pi in enumerate(p):
                binary_domain = domain < pi
                label, num = scpi.measurements.label(binary_domain)
                perc_x = np.intersect1d(label[0, :], label[-1, :])
                percx = perc_x[np.where(perc_x > 0)]
                if len(percx) > 0:
                    Ni[j, k] += 1
        Pi[:, k] = Ni[:, k] / N # For all p we store values for each
    
    return Pi


P = P_func(p, L, N)
Pi = Pi_func(p, L, N)
```

```python
for i, l in enumerate(L):
    plt.plot(p, P[:, i], label=f'{l}')
plt.legend()
plt.xlabel('p')
plt.ylabel('$P(p, L)$')
plt.show()

for i, l in enumerate(L):
    plt.plot(p, Pi[:, i], label=f'{l}')
plt.legend()
plt.xlabel('p')
plt.ylabel('$\\Pi(p, L)$')
plt.show()
```

For smÃ¥ systemer, $L = 1, 2$ kan vi regne ut den faktiske verdien for perkolasjonssannsynligheten. For $L=2$ har vi 16 forskjellige kombinasjoner som vist ovenfor. Ã˜ker vi derimot $L=3$ fÃ¥r vi $2^{L^2} = 2^9 = 512$ forskjellige kombinasjoner. For et stÃ¸rre system mÃ¥ vi derfor gjÃ¸re en tilnÃ¦rming ved hjelp av monte-carlo simulasjon over forskjellige system. $\Pi(p_i, L) \approx N_i / N$, hvor $N$ er antall MC simulasjoner og $N_i$ er antall system som perkolerer. Vi setter opp eksempelvis 1000 forskjellige system for en gitt p og L, sÃ¥ finner vi ut hvor mange av disse systemene som perkolerer. Desto flere simulasjoner vi gjÃ¸r, desto bedre blir approksimasjonen vÃ¥r. 

Sannsynligheten for at et site er en del av et spanning cluster finner vi fra $P(p, L) = M_s / L^2$, hvor $M_s$ er massen til et cluster av stÃ¸rrelse s. Vi gjÃ¸r igjen monte-carlo simulasjon over forskjellige lattice av stÃ¸rrelse L med en gitt p. NÃ¥r vi har et spanning cluster finner vi massen til clusteret ved Ã¥ summere antall sites i clusteret. Etter endt MC simulasjon finner vi $P(p, L) \approx M_s / (N L^2)$ hvor N er antall MC simulasjoner.
