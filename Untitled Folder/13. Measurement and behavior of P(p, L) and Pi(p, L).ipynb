### Discuss the behavior of $P(p, L)$ and $\Pi(p, L)$ in a finite system of size L

$\Pi(p,L)$ angir sannsynligheten for at systemet perkolerer. Når $p \rightarrow p_c$ fra 0 øker perkolasjonssannsynligheten. Hvor skarpt perkolasjonssannsynligheten øker avhenger av systemstørrelsen L. For $p < p_c$ for små system ser ser vi at sannsynligheten for perkolasjon øker raskere enn for store system. For $p > p_c$ ser vi at sannsynligheten for perkolasjon synker med system størrelsen. $Pi(p,L)$ ut til å gå mot en step-funksjon når vi øker L. 

$P(p,L)$ angir sannsynligheten for at et site tilhører et spanning cluster. Når $p \rightarrow p_c$ øker $P(p,L)$, hvor sannsynligheten avhenger av systemstørrelsen. Vi observerer samme oppførsel som for $Pi(p,L)$ mtp systemstørrelse og sannsynlighet rundt $p_c$. Vi ser at $P(p, L)$ øker raskt like etter $p_c$, årsaken til dette kan være at flere små clustere nå blir koblet sammen, og vi ser en rask økning i clusterstørrelse, og følgelig er sannsynligheten for at et spesifikt site er en del av et spanning cluster. En stund etter dette ser vi at $P(p,L)$ øker omtrent lineært, som kan være et resultat av at det kun er noen sites som gjør at systemet perkolerer.

### Måle $P(p,L)$ og $\Pi(p,L)$

Density of the spanning cluster er gitt ved

\begin{equation}
    P(p,L) = \frac{M_s}{L^d}
\end{equation}

hvor $M_s$ er massen til spanning cluster, L er systemstørrelsen og d er dimensjonen til systemet. Vi måler $P(p,L)$ ved monte-carlo simulering. Vi genererer et tilfeldig system for hver MC simulasjon og sjekker om dette perkolerer, dersom det perkolerer lagrer vi massen til spanning clusteret ved å telle antall sites. Når simulasjonen er ferdig finner vi 

\begin{equation}
P = \frac{M_{s,tot}}{N L^d}
\end{equation}

Perkolasjonssannsynligheten for større systemet er gitt ved

\begin{equation}
\Pi(p,L) \approx \frac{N_i}{N}
\end{equation}

hvor $N$ er antall MC simulasjoner og $N_i$ er antall system som perkolerer.  Vi setter opp eksempelvis 1000 forskjellige system for en gitt p og L, vi teller deretter hvor mange av disse systemene som har en cluster som strekker seg fra den ene siden til den andre, altså at systemet perkolerer. Desto flere simulasjoner vi gjør, desto bedre blir approksimasjonen vår.

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import scipy.ndimage as scpi

L = np.array([25, 50, 100, 200, 400, 800])
p = np.linspace(0.3, 0.9, 51)
N = 1000

def func(p, L, N):
    nL = len(L)
    nx = len(p)

    Ni = np.zeros((nx, nL))
    Pi = np.zeros((nx, nL))
    P  = np.zeros((nx, nL))

    for k, l in tqdm(enumerate(L)):
        for i in range(N):
            domain = np.random.rand(l, l)
            for j, pi in enumerate(p):
                binary_domain = domain < pi
                label, num = scpi.measurements.label(binary_domain)
                perc_x = np.intersect1d(label[0, :], label[-1, :])
                percx = perc_x[np.where(perc_x > 0)]
                if len(percx) > 0:# or len(percy) > 0:
                    Ni[j, k] += 1
                    area = scpi.measurements.sum(binary_domain, label, percx[0])
                    P[j, k] += area
        Pi[:, k] = Ni[:, k] / N # For all p we store values for each
        P[:, k] /= (N * l * l)
    return Pi, P
```

```python
Pi, P = func(p, L, N)
```

```python
plt.figure(figsize=(12,6))
for i,l in enumerate(L):
    plt.plot(p, Pi[:, i], label=f'L={l}')
plt.legend()
plt.xlabel('p')
plt.ylabel('$\Pi(p,L)$')
plt.show()

plt.figure(figsize=(12,6))
for i,l in enumerate(L):
    plt.plot(p, P[:, i], label=f'L={l}')
plt.legend()
plt.xlabel('p')
plt.ylabel('$P(p,L)$')
plt.show()
```
