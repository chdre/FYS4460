### $\Pi(p,L)$

$\Pi(p,L)$ angir sannsynligheten for at systemet perkolerer. Når $p \rightarrow p_c$ fra 0 øker perkolasjonssannsynligheten. Hvor skarpt perkolasjonssannsynligheten øker avhenger av systemstørrelsen L. For $p < p_c$ for små system ser ser vi at sannsynligheten for perkolasjon øker raskere enn for store system. For $p > p_c$ ser vi at sannsynligheten for perkolasjon synker med system størrelsen. $Pi(p,L)$ ut til å gå mot en step-funksjon når vi øker L. 

Perkolasjonssannsynligheten er gitt ved

\begin{equation}
\Pi(p,L) \approx \frac{N_i}{N}
\end{equation}

hvor $N$ er antall MC simulasjoner og $N_i$ er antall system som perkolerer.  Vi setter opp eksempelvis 1000 forskjellige system for en gitt p og L, vi teller deretter hvor mange av disse systemene som har en cluster som strekker seg fra den ene siden til den andre, altså at systemet perkolerer. Desto flere simulasjoner vi gjør, desto bedre blir approksimasjonen vår.

### Finite size scaling

Vi ser at $\Pi(p_c, L)$ ikke går mot 0 eller divergerer, følgelig kan den ikke være en funksjon av kun $\xi = \xi_0|p - p_c|^{-\nu} \rightarrow 0$ når $p \rightarrow p_c$. Vi foreslår istedenfor at den følger 

\begin{equation}
    \Pi (p,L) = \xi^0 f\left(\frac{L}{\xi}\right).
\end{equation}

Setter inn for $\xi$

\begin{equation}
    \Pi(p, L) = f\left(L\xi_0|p - p_c|^\nu\right)
\end{equation}

Flytter eksponenten inn og får

\begin{equation}
\Pi(p,L) = f\left(\xi_0\left(L^{1/\nu}(p-p_c)\right)^\nu\right)
\end{equation}

Vi definerer nå skaleringsfunksjonen $\Phi(u) = f(\xi_0 u^{1/\nu})$ og får

\begin{equation}
    \Pi(p,L) = \Phi\left(L^{1/\nu}(p - p_c)\right)
\end{equation}

### Estimere $p_c$

For å estimere en verdi for $p_c$ velger en verdi $p_x$ og bruker denne til å finne $\Pi(p_x,L) = x$ og setter dette inn i skalerings ansatzen ovenfor, slik at vi får

\begin{equation}
    \Pi(p_x, L) = x = \Phi\left(L^{1/\nu}(p - p_c)\right).
\end{equation}



Vi tar så den inverse av $\Phi$

\begin{equation}
    L^{1/\nu}(p_x - p_c) = \Phi^{-1}(x) = C_x,
\end{equation}

hvor $C_x$ er en konstant. Multipliserer vi med $L^{-1/\nu}$ får vi

\begin{equation}
    p_x - p_c = L^{-1/\nu}C_x
\end{equation}

hvor vi nå ser at $- p_c$ vil være intercepten til den lineære tilpasningen til $p_x$ mot $L^{-1/\nu}$. Ettersom $C_x$ er en konstant kan vi se bort ifra denne.


I prosjekt 3 gjorde vi dette for $p_{x_1} = 0.3$ og $p_{x_2} = 0.8$

Som resulterte i $p_c$ for $x_1$: 0.59293 og x$_2$: 0.59251. 

### Estimere $\nu$

Når vi skal estimere $\nu$ kan vi se på endringen i p for to verdier for $x = x_1, x_2$

\begin{equation}
    dp = p_{\Pi=x_1}(L) - p_{\Pi=x_2}(L) = \left(C_{x_1} - C_{x_2}\right)L^{-1/\nu} = C_xL^{-1/\nu}
\end{equation}

hvor vi har gruppert $C_{x_1} - C_{x_2} = C_x$, denne kan som ovenfor fjernes ettersom den kun skalerer løsningen. 

Gjør vi nå en lineærtilpasning på $\log(p_{x_1} - p_{x_2})$ mot $\log(L^{-1/\nu}) = \log(L^{y}) = y\log(L)$ ser vi at $y = -1/\nu$ er stigningstallet til tilpasningen. I prosjekt 3 gjorde vi dette for samme $x_1$ og $x_2$ som over.

Dette ga $\nu = 1.368$, sammenliknet med teoretisk verdi $4/3$ er dette ganske godt.

Vi ser at finite size scaling teori gir ganske gode resultater til tross for at systemene er relativt små.
