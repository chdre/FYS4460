```python
import numba
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import time
```

```python
@numba.njit()
def percwalk(cluster, max_steps):
    walker_map = np.zeros((2, max_steps), dtype=np.int16)
    displacement = np.zeros((2, max_steps), dtype=np.int16)
    directions = np.zeros((2, 4), dtype=np.int16)
    neighbor_arr = np.zeros(4, dtype=np.int16)

    directions[0, 0] = 1
    directions[1, 0] = 0
    directions[0, 1] = -1
    directions[1, 1] = 0
    directions[0, 2] = 0
    directions[1, 2] = 1
    directions[0, 3] = 0
    directions[1, 3] = -1

    Lx, Ly = cluster.shape
    ix = np.random.randint(Lx)
    iy = np.random.randint(Ly)
    walker_map[0, 0] = ix
    walker_map[1, 0] = iy
    step = 0

    if not cluster[ix, iy]:
        return walker_map, displacement, step
    while step < max_steps - 1:
        neighbor = 0
        for idir in numba.prange(directions.shape[1]):
            dr = directions[:, idir]
            iix = ix + dr[0]
            iiy = iy + dr[1]
            
            if iix >= Lx:
                iix -= Lx
            if iix < 0:
                iix += Lx
            if iiy >= Ly:
                iiy -= Ly
            if iiy < 0:
                iiy += Ly
            if cluster[iix, iiy]:
                neighbor_arr[neighbor] = idir
                neighbor += 1

        if neighbor == 0:
            return walker_map, displacement, step
        
        randdir = np.random.randint(neighbor)
        dirr = neighbor_arr[randdir]
        ix += directions[0, dirr]
        iy += directions[1, dirr]
        step += 1
        if ix >= Lx:
            ix -= Lx
        if ix < 0:
            ix += Lx
        if iy >= Ly:
            iy -= Ly
        if iy < 0:
            iy += Ly
        walker_map[0, step] = ix
        walker_map[1, step] = iy
        displacement[:, step] = displacement[:, step-1] + directions[:, dirr]
    
    return walker_map, displacement, step
```

```python
@numba.njit()
def percwalk_nonperiodic(cluster, max_steps):
    walker_map = np.zeros((2, max_steps), dtype=np.int16)
    displacement = np.zeros((2, max_steps), dtype=np.int16)
    directions = np.zeros((2, 4), dtype=np.int16)
    neighbor_arr = np.zeros(4, dtype=np.int16)
    
    directions[0, 0] = 1
    directions[1, 0] = 0
    directions[0, 1] = -1
    directions[1, 1] = 0
    directions[0, 2] = 0
    directions[1, 2] = 1
    directions[0, 3] = 0
    directions[1, 3] = -1
    
    # Initial random position
    Lx, Ly = cluster.shape
    ix = np.random.randint(Lx)
    iy = np.random.randint(Ly)
    walker_map[0, 0] = ix
    walker_map[1, 0] = iy
    step = 0
    
    if not cluster[ix, iy]: # Landed outside the cluster
        return walker_map, displacement, step
    
    while step < max_steps-1:
        # Make list of possible moves
        neighbor = 0
        for idir in range(directions.shape[1]):
            dr = directions[:, idir]
            iix = ix + dr[0]
            iiy = iy + dr[1]
            if 0 <= iix < Lx and 0 <= iiy < Ly and cluster[iix, iiy]:
                neighbor_arr[neighbor] = idir
                neighbor += 1
        if neighbor == 0: # No way out, return
            return walker_map, displacement, step
    
        # Select random direction from 0 to neighbor-1
        randdir = np.random.randint(neighbor)
        dir = neighbor_arr[randdir]
        ix += directions[0, dir]
        iy += directions[1, dir]
        step += 1
        walker_map[0, step] = ix
        walker_map[1, step] = iy
        displacement[:, step] = displacement[:,step-1]+directions[:, dir]
        
    return walker_map, displacement, step
```

```python
L = 100
pc = 0.5927
binary_domain = np.random.rand(L, L) < pc
plt.imshow(binary_domain)
N = 10000

for i in range(30):
    walker_map, displacement, steps = percwalk(binary_domain, N)
    plt.plot(walker_map[1, :steps], walker_map[0, :steps], 'o', markersize=1)
```

```python
@numba.njit()
def find_displacements(p, L, num_systems, num_walkers, max_steps):
    displacements = np.zeros(max_steps, dtype=np.float64)
    for sys in numba.prange(num_systems):
        binary_domain = np.random.rand(L, L) < p
        for rw in numba.prange(num_walkers):
            num_steps = 0
            while num_steps <= 1:
                walker_map, displacement, num_steps = percwalk(binary_domain, max_steps)
            displacements += np.sum(displacement * displacement, axis=0)
    displacements /= (num_walkers * num_systems)
    
    return displacements
```

### Diffusjon for $p > p_c$
Når okkupasjonssannsynligheten p er 1 er systemet homogent og virrevandreren kan gå fritt over systemet, og vi kan da beskrive diffusjonen som $\langle r^2 \rangle = D(1)t$. Når vi gjør p mindre forventer vi at diffusjonen går tregere. Fra Einstein relasjonen for diffusjon har vi at $D(p) \propto (p - p_c)^\mu$

```python
p2 = np.linspace(0.6, 0.75, 21)
L = 500
N = 10000
num_walkers = 1000
num_systems = 100

start = time.time()
@numba.njit(parallel=True)
def func2():
    disps2 = np.zeros((len(p2), N - 1), dtype=np.float64)
    for i in numba.prange(len(p2)):
        pi = p2[i]
        disps2[i, :] = find_displacements(pi, L, num_systems, num_walkers, N)[1:]
    return disps2
disps2 = func2()
print(f'Time: {time.time() - start}')
```

```python
for i, d in enumerate(disps2):
    t = np.arange(len(d))[1:]
    plt.plot(np.log(t), np.log(d[1:]), label=f'{p2[i]:.3f}')
# plt.legend()
plt.xlabel('$\log(t)$')
plt.ylabel('$\log(r^2)$')
plt.show()
```

```python
D = []
for i, d in enumerate(disps2):
    t = np.arange(len(d))
    idx = np.argwhere(np.log(t[1:]) > 2)[:, 0]
    t = t[idx]
    d = d[idx]
    slope, _ = np.polyfit(t, d, 1)
    D.append(slope)
```

```python
plt.plot(np.log((p2 - pc)), np.log(D), '-o')
plt.show()
```

```python
D = np.array(D)
idx = np.argwhere(np.log(p2 - pc) > -3.0)[:, 0]
# x = np.log(p2 - pc)[idx]
# idx2 = np.argwhere(x < -1.5)
# x = x[idx2][:, 0]
# y = D[idx]
# y = y[idx2][:, 0]

x = np.log(p2 - pc)[idx]
y = np.log(D)[idx]

mu_p2, _ = np.polyfit(x, y, 1)
print(mu_p2)
```

Ser at $\mu \approx 0.90$, som er mindre enn teoretiske verdien $1.3$.

Når t er mindre enn en cross-over tid $t_0$ ser vi at oppførselen til virrevandreren følger en power-law med eksponent $2k$. Virrevandreren har før $t_0$ ikke blitt "kjent" med den endelige størrelsen på clusteret, og oppførselen er lik den for et uendelig cluster. Etter cross-over tiden er virrevandreren kjent med at clusteret har en endelig størrelse, og oppførselen likner den for diffusjon med diffusjonskonstant $D(p)$ for $p > p_c$ eller så når den et konstant platå for $p < p_c$.

Vi bruker en scaling ansatz for  $\langle r^2 \rangle = t^{2k} f[(p - p_c)t^x]$

```python
beta = 5 / 36
nu = 4 / 3
mu = 1.3
k = 0.5 * (1 - mu / (2 * nu + mu - beta))
x = 1 / (2 * nu + mu - beta)

for i, r2 in enumerate(disps):
    t = np.arange(len(r2))[1:]
    xx = r2[1:] * t ** (- 2 * k)
    yy = (pc - p1[i]) * t ** x
    plt.plot(yy, xx, 'o', markersize=1)
plt.show()
```
