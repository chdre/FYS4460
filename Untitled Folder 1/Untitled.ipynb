```python
import numba
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import time

plt.rcParams['figure.figsize'] = (8.0, 6.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# Set fontsizes in figures
params = {'axes.labelsize': 'x-large',
          'axes.titlesize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large',
          'legend.fontsize': 'large',
          'legend.handlelength': 2}
plt.rcParams.update(params)
```

```python
@numba.njit()
def percwalk(cluster, max_steps):
    walker_map = np.zeros((2, max_steps), dtype=np.int16)
    displacement = np.zeros((2, max_steps), dtype=np.int16)
    directions = np.zeros((2, 4), dtype=np.int16)
    neighbor_arr = np.zeros(4, dtype=np.int16)

    directions[0, 0] = 1
    directions[1, 0] = 0
    directions[0, 1] = -1
    directions[1, 1] = 0
    directions[0, 2] = 0
    directions[1, 2] = 1
    directions[0, 3] = 0
    directions[1, 3] = -1

    Lx, Ly = cluster.shape
    ix = np.random.randint(Lx)
    iy = np.random.randint(Ly)
    walker_map[0, 0] = ix
    walker_map[1, 0] = iy
    step = 0

    if not cluster[ix, iy]:
        return walker_map, displacement, step
    while step < max_steps - 1:
        neighbor = 0
        for idir in numba.prange(directions.shape[1]):
            dr = directions[:, idir]
            iix = ix + dr[0]
            iiy = iy + dr[1]

            if iix >= Lx:
                iix -= Lx
            if iix < 0:
                iix += Lx
            if iiy >= Ly:
                iiy -= Ly
            if iiy < 0:
                iiy += Ly
            if cluster[iix, iiy]:
                neighbor_arr[neighbor] = idir
                neighbor += 1

        if neighbor == 0:
            return walker_map, displacement, step

        randdir = np.random.randint(neighbor)
        dirr = neighbor_arr[randdir]
        ix += directions[0, dirr]
        iy += directions[1, dirr]
        step += 1
        if ix >= Lx:
            ix -= Lx
        if ix < 0:
            ix += Lx
        if iy >= Ly:
            iy -= Ly
        if iy < 0:
            iy += Ly
        walker_map[0, step] = ix
        walker_map[1, step] = iy
        displacement[:, step] = displacement[:, step-1] + directions[:, dirr]

    return walker_map, displacement, step
```

```python
L = 100
pc = 0.5927
binary_domain = np.random.rand(L, L) < pc
plt.imshow(binary_domain)
N = 10000

for i in range(30):
    walker_map, displacement, steps = percwalk(binary_domain, N)
    plt.plot(walker_map[1, :steps], walker_map[0, :steps], 'o', markersize=1)
```

```python
@numba.njit()
def find_displacements(p, L, num_systems, num_walkers, max_steps):
    displacements = np.zeros(max_steps, dtype=np.float64)
    for sys in numba.prange(num_systems):
        binary_domain = np.random.rand(L, L) < p
        for rw in numba.prange(num_walkers):
            num_steps = 0
            while num_steps <= 1:
                walker_map, displacement, num_steps = percwalk(binary_domain, max_steps)
            displacements += np.sum(displacement * displacement, axis=0)
    displacements /= (num_walkers * num_systems)
    
    return displacements
```

Nr t er mindre enn en cross-over tid $t_0$ ser vi at oppfrselen til virrevandreren flger en power-law med eksponent $2k$, $\langle r^2 \rangle \propto t^{2k}$, uavhengig av p.

Diffusjon for $p = p_c$
Nr okkupasjonssannsynligheten p er 1 er systemet homogent og virrevandreren kan g fritt over systemet, og vi kan da beskrive diffusjonen som $\langle r^2 \rangle = D(1)t$.

For $p < p_c$ har virrevandreren fr $t_0$ ikke blitt "kjent" med den endelige strrelsen p clusteret, og oppfrselen er lik den for et uendelig cluster. Etter cross-over tiden er virrevandreren kjent med at clusteret har en endelig strrelse, og $r^2$ nr et konstant plat, som svarer til at virrevandreren har blitt kjent med clusterstrrelsen og ftt sin bevegelse begrenset. Vi ser da at forflytningen $r^2 \propto R_s^2$, alts forflytningen er proposjonal med gyrasjonsradiusen for en cluster av strrelse s. Vi ser at jo nrmere $p$ er $p_c$, desto lengre er cross-over tiden, alts; jo lengre tid tar det fr virrevandreren oppdager at den er stengt inne i et cluster. 

For $p > p_c$: Vi ser at nr vi ker p ker ogs $r^2$, alts diffunderer systemet raskere. Intuitivt gir dette mening, ettersom okkupasjonssannsynligheten p (som svarer til porositeten til systemet) er strre; virrevandreren har mer plass  boltre seg p. Vi ser en liner oppfrsel $\langle r^2 \rangle = D(p)t$ som likner den for diffusjon med diffusjonskonstant $D(p)$ for $p > p_c$. Fra Einstein relasjonen for diffusjon har vi at $D(p) \propto (p - p_c)^\mu$

```python
p2 = np.linspace(0.6, 0.75, 21)
L = 500
N = 10000
num_walkers = 1000
num_systems = 100

start = time.time()


@numba.njit(parallel=True)
def func2():
    disps2 = np.zeros((len(p2), N - 1), dtype=np.float64)
    for i in numba.prange(len(p2)):
        pi = p2[i]
        disps2[i, :] = find_displacements(
            pi, L, num_systems, num_walkers, N)[1:]
    return disps2


disps2 = func2()
print(f'Time: {time.time() - start}')
```

### Diffusjonskonstanten og $\nu$
For $p > p_c$ har vi ved Einsteinrelasjonen at $D(p) \propto (p - p_c)^\mu$. Vi finner diffisjonskonstanten fra $\langle r^2 \rangle = D(p) t$ ved  finne stigningstallet til $\langle r^2(t)\rangle$ for forskjellige p. Deretter finner vi stigningstallet til den linere delen av logaritmiske plottet.

```python
D = []
for i, r2 in enumerate(disps2):
    t = np.arange(len(r2))
    idx = np.argwhere(np.log(t[1:]) > 2)[:, 0]
    t = t[idx]
    r2 = r2[idx]
    Dp, _ = np.polyfit(t, r2, 1)
    D.append(Dp)
```

```python
plt.plot(np.log((p2 - pc)), np.log(D), '-o')
plt.xlabel('$\\log(p - p_c)$')
plt.ylabel('$\\log(D(p))$')
plt.title('Diffusjonskonstanten for $p > p_c$')
plt.show()
```

```python
D = np.array(D)
idx = np.argwhere(np.log(p2 - pc) > -3.0)[:, 0]
x = np.log(p2 - pc)[idx]
y = np.log(D)[idx]

mu_p2, _ = np.polyfit(x, y, 1)
print(mu_p2)
```

Ser at $\mu \approx 0.90$, som er mindre enn teoretiske verdien, for 2D, $1.3$.

### Tester scaling ansatz
Vi bruker scaling ansatz: $\langle r^2 \rangle = t^{2k} f[(p - p_c)t^x]$

Vi antar at scaling funksjonen f(u) har flgende form:

$f(u) = \text{constant}$ if $|u| << 1$

$f(u) = u^{\mu}$ if $u >> 1$

$f(u) = (-u)^{\beta - 2\nu}$ if $u << -1$

For $p > p_c$ har vi da at $\langle r^2 \rangle \propto (p - p_c)^{\mu}t$, setter vi dette inn i scaling ansatzen fr vi $\langle r^2 \rangle \propto t^{2k}((p - p_c)t^x)^{\mu}$, som gir at $k = (1 - \mu x)/2$.

For $p < p_c$ har vi at $\langle r^2 \rangle \propto (p - p_c)^{\beta - 2\nu}$, scaling ansatzen gir $\langle r^2 \rangle \propto t^{2k}((p_c - p)t^x)^{\beta - 2\nu}$. Dette gir at $k = (1 - \mu/(2\nu + \mu - \beta))/2$ og $x = 1/(2\nu + \mu - \beta)$. 

```python
beta = 5 / 36
nu = 4 / 3
mu = 1.3
k = 0.5 * (1 - mu / (2 * nu + mu - beta))
x = 1 / (2 * nu + mu - beta)

for i, r2 in enumerate(disps):
    t = np.arange(len(r2))[1:]
    xx = r2[1:] * t ** (- 2 * k)
    yy = (pc - p1[i]) * t ** x
    plt.plot(yy, xx)  # , markersize=1)
plt.xlabel('$(p_c - p)t^x$')
plt.ylabel('$\\langle r^2 \rangle t^{-2k}$')
plt.title('$Data collapse for p < p_c$')
plt.show()
```
