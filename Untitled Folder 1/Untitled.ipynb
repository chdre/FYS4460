# Introduksjon

I dette prosjektet har vi studert diffusjon på et nanoporøst system. Vi lager et nanoporøst system med en okkupasjonssannsynlighet p, som svarer til porositeten $\phi = V_p/V_{tot}$ dersom vi anser okkuperte sites som hulrom i systemet. Ved å slippe en virrevandrer i systemet og la den bevege seg på et cluster kan vi studere diffusjon ved å se på mean squared displacement til virrevandreren ved hjelp av Monte-Carlo simulasjon. 

```python
import numba
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import time

plt.rcParams['figure.figsize'] = (8.0, 6.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# Set fontsizes in figures
params = {'axes.labelsize': 'x-large',
          'axes.titlesize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large',
          'legend.fontsize': 'large',
          'legend.handlelength': 2}
plt.rcParams.update(params)
```

```python
@numba.njit()
def percwalk(cluster, max_steps):
    walker_map = np.zeros((2, max_steps), dtype=np.int16)
    displacement = np.zeros((2, max_steps), dtype=np.int16)
    directions = np.zeros((2, 4), dtype=np.int16)
    neighbor_arr = np.zeros(4, dtype=np.int16)

    directions[0, 0] = 1
    directions[1, 0] = 0
    directions[0, 1] = -1
    directions[1, 1] = 0
    directions[0, 2] = 0
    directions[1, 2] = 1
    directions[0, 3] = 0
    directions[1, 3] = -1

    Lx, Ly = cluster.shape
    ix = np.random.randint(Lx)
    iy = np.random.randint(Ly)
    walker_map[0, 0] = ix
    walker_map[1, 0] = iy
    step = 0

    if not cluster[ix, iy]:
        return walker_map, displacement, step
    while step < max_steps - 1:
        neighbor = 0
        for idir in numba.prange(directions.shape[1]):
            dr = directions[:, idir]
            iix = ix + dr[0]
            iiy = iy + dr[1]

            if iix >= Lx:
                iix -= Lx
            if iix < 0:
                iix += Lx
            if iiy >= Ly:
                iiy -= Ly
            if iiy < 0:
                iiy += Ly
            if cluster[iix, iiy]:
                neighbor_arr[neighbor] = idir
                neighbor += 1

        if neighbor == 0:
            return walker_map, displacement, step

        randdir = np.random.randint(neighbor)
        dirr = neighbor_arr[randdir]
        ix += directions[0, dirr]
        iy += directions[1, dirr]
        step += 1
        if ix >= Lx:
            ix -= Lx
        if ix < 0:
            ix += Lx
        if iy >= Ly:
            iy -= Ly
        if iy < 0:
            iy += Ly
        walker_map[0, step] = ix
        walker_map[1, step] = iy
        displacement[:, step] = displacement[:, step-1] + directions[:, dirr]

    return walker_map, displacement, step
```

```python
L = 100
pc = 0.59275
binary_domain = np.random.rand(L, L) < pc
plt.imshow(binary_domain)
N = 10000

for i in range(30):
    walker_map, displacement, steps = percwalk(binary_domain, N)
    plt.plot(walker_map[1, :steps], walker_map[0, :steps], 'o', markersize=1)
```

```python
@numba.njit()
def find_displacements(p, L, num_systems, num_walkers, max_steps):
    displacements = np.zeros(max_steps, dtype=np.float64)
    for sys in numba.prange(num_systems):
        binary_domain = np.random.rand(L, L) < p
        for rw in numba.prange(num_walkers):
            num_steps = 0
            while num_steps <= 1:
                walker_map, displacement, num_steps = percwalk(binary_domain, max_steps)
            displacements += np.sum(displacement * displacement, axis=0)
    displacements /= (num_walkers * num_systems)
    
    return displacements
```

Når t er mindre enn en cross-over tid $t_0$ ser vi at oppførselen til virrevandreren følger en power-law med eksponent $2k$, $\langle r^2 \rangle \propto t^{2k}$, uavhengig av p.

Diffusjon for $p = p_c$
Når okkupasjonssannsynligheten p er 1 er systemet homogent og virrevandreren kan gå fritt over systemet, og vi kan da beskrive diffusjonen som $\langle r^2 \rangle = D(p=1)t$.

# Diffusjon for $p > p_c$

```python
p2 = np.linspace(0.6, 0.75, 21)
L = 500
N = 10000
num_walkers = 1000
num_systems = 100

start = time.time()


@numba.njit(parallel=True)
def func2():
    disps2 = np.zeros((len(p2), N - 1), dtype=np.float64)
    for i in numba.prange(len(p2)):
        pi = p2[i]
        disps2[i, :] = find_displacements(
            pi, L, num_systems, num_walkers, N)[1:]
    return disps2


disps2 = func2()
print(f'Time: {time.time() - start}')
```

```python
for i, d in enumerate(disps2):
    t = np.arange(len(d))[1:]
    plt.plot(np.log(t), np.log(d[1:]), label=f'{p2[i]:.3f}')
# plt.legend()
plt.xlabel('$\log(t)$')
plt.ylabel('$\log(r^2)$')
plt.show()
```

For $p > p_c$: Vi ser at når vi øker p øker også $r^2$, altså diffunderer systemet raskere. Intuitivt gir dette mening, ettersom okkupasjonssannsynligheten p er større; virrevandreren har mer plass å boltre seg på. Vi ser en lineær oppførsel $\langle r^2 \rangle = D(p)t$ etter cross-overtiden som forteller oss at vi systemet diffunderer normalt med diffusjonskonstant $D(p)$ for $p > p_c$. For cross-overtiden ser vi anomal diffusjon. 

## Diffusjon for $p < p_c$

```python
p1 = np.linspace(0.41, 0.59, 11)
L = 500
N = 10000
num_walkers = 1000
num_systems = 100

start = time.time()


@numba.njit(parallel=True)
def func1():
    disp = np.zeros((len(p1), N - 1), dtype=np.float64)
    for i in numba.prange(len(p1)):
        pi = p1[i]
        disp[i, :] = find_displacements(pi, L, num_systems, num_walkers, N)[1:]
    return disp


disps = func1()
print(f'Time: {time.time() - start}')
```

```python
for r2 in disps:
    t = np.arange(len(r2))[1:]
    plt.plot(np.log(t), np.log(r2[1:]))
# plt.legend(p1)
plt.xlabel('$\log(t)$')
plt.ylabel('$\log(r^2)$')
plt.show()
```

For $p < p_c$ har virrevandreren før $t_0$ ikke blitt "kjent" med den endelige størrelsen på clusteret, og oppførselen er lik den for et uendelig cluster. Etter cross-over tiden er virrevandreren kjent med at clusteret har en endelig størrelse, og $r^2$ når et konstant platå, som svarer til at virrevandreren har blitt kjent med clusterstørrelsen og fått sin bevegelse begrenset. Vi ser da at forflytningen $r^2 \propto R_s^2$, altså forflytningen er proposjonal med gyrasjonsradiusen for et cluster av størrelse s. Vi ser at jo nærmere $p$ er $p_c$, desto lengre er cross-over tiden, altså; jo lengre tid tar det før virrevandreren oppdager at den er stengt inne i et cluster. 

## Finner numerisk verdi for k
For $p=p_c$ går diffusjonen som $\langle r^2\rangle \propto t^{2k}$. Ved å se på stigningstallet til logaritmen av $\langle r^2(t)\rangle$ finner vi k ved lineærtilpasning av

\begin{equation}
\log(\langle r^2 \rangle) \propto \log(t^{2k}) \Rightarrow \log(\langle r^2\rangle) \propto 2k \log(t),
\end{equation}

hvor stigningstallet er $2k$.

```python
L = 500
num_walkers = 500
num_systems = 100
N = 10000
r2 = find_displacements(pc, L, num_systems, num_walkers, N)[1:]
```

```python
t = np.arange(len(r2))

plt.loglog(t[1:], r2[1:], label='p=0.5927')
plt.title('$k = \\frac{d\\log(r^2)}{d\\log(t)}$')
plt.xlabel('$\log(t)$')
plt.ylabel('$\log(r^2)$')
plt.legend()
plt.show()
```

```python
k_numerical, _ = np.polyfit(np.log(t[1:]), np.log(r2[1:]), 1)
print(f'k={k_numerical / 2:.3f}')
```

Vi får $k \approx 0.341$, sammenliknet med teoretisk verdi $k \approx 0.33$.

## Diffusjonskonstanten $p > p_c$
Vi finner diffisjonskonstanten fra $\langle r^2 \rangle = D(p) t$ ved å finne stigningstallet til $\langle r^2(t)\rangle$ for forskjellige p.

```python
D = []
for i, r2 in enumerate(disps2):
    t = np.arange(len(r2))
    idx = np.argwhere(np.log(t[1:]) > 2)[:, 0]
    t = t[idx]
    r2 = r2[idx]
    Dp, _ = np.polyfit(t, r2, 1)
    D.append(Dp)
```

```python
plt.plot(np.log((p2 - pc)), np.log(D), '-o')
plt.xlabel('$\\log(p - p_c)$')
plt.ylabel('$\\log(D(p))$')
plt.title('Diffusjonskonstanten for $p > p_c$')
plt.show()
```

## Finne $\mu$
For $p > p_c$ har vi ved Einsteinrelasjonen for diffusjon at $D(p) \propto (p - p_c)^\mu$. For å finne $\mu$ kan vi gjøre en lineærtilpasning i logaritmisk skala, slik som tidligere. 

\begin{equation}
\log(D(p)) \propto \mu \log(p - p_c)
\end{equation}

```python
D = np.array(D)
idx = np.argwhere(np.log(p2 - pc) > -3.0)[:, 0]
x = np.log(p2 - pc)[idx]
y = np.log(D)[idx]

mu_p2, _ = np.polyfit(x, y, 1)
print(mu_p2)
```

Ser at $\mu \approx 0.90$, som er mindre enn teoretiske verdien, for 2D, $1.3$.

## Scaling ansatz
Vi bruker scaling ansatz: $\langle r^2 \rangle = t^{2k} f[(p - p_c)t^x]$

Vi antar at scaling funksjonen f(u) har følgende form:

\begin{align}
 &\text{constant} \quad |u| << 1\\
f(u) = & u^{\mu} \quad\quad\quad u >> 1 \\
&(-u)^{\beta - 2\nu} \quad u << -1
\end{align}

For $\mathbf{p > p_c}$ har vi da

\begin{equation}
\langle r^2 \rangle \propto (p - p_c)^{\mu}t,
\end{equation}

setter vi dette inn i scaling ansatzen får vi

\begin{equation}
\langle r^2 \rangle \propto t^{2k}((p - p_c)t^x)^{\mu},
\end{equation}

som gir at 

\begin{equation}
k = \frac{1 - \mu x}{2}
\end{equation}


For $\mathbf{p < p_c}$ har vi at 

\begin{equation}
\langle r^2 \rangle \propto (p - p_c)^{\beta - 2\nu}
\end{equation}

scaling ansatzen gir

\begin{equation}
\langle r^2 \rangle \propto t^{2k}((p_c - p)t^x)^{\beta - 2\nu}.
\end{equation}

Dette gir

\begin{align}
k &= \frac{1}{2}(1 - \frac{\mu}{(2\nu + \mu - \beta)})\\
x &= \frac{1}{(2\nu + \mu - \beta)},
\end{align}

som gjør at vi kan bestemme k og x direkte og sette sistnevnte inn i likningen for k for $p > p_c$.

Vi kan nå teste scaling ansatzen.

\begin{equation}
\langle r^2 \rangle = t^{2k} f[(p - p_c)t^x]
\end{equation}

hvor vi multipliserer med $t^{-2k}$ og får

\begin{equation}.
\langle r^2 \rangle t^{-2k} =  f[(p - p_c)t^x].
\end{equation}

Dersom scaling ansatzen vår er gyldig forventer vi at

\begin{equation}
\langle r^2 \rangle t^{-2k} =  (p - p_c)t^x
\end{equation}

for p sammenfaller til lik oppførsel; at de kollapser til funksjonen $f(u)$.

```python
beta = 5 / 36
nu = 4 / 3
mu = 1.3
k = 0.5 * (1 - mu / (2 * nu + mu - beta))
x = 1 / (2 * nu + mu - beta)

for i, r2 in enumerate(disps):
    t = np.arange(len(r2))[1:]
    yy = r2[1:] * t ** (- 2 * k)
    xx = (p1[i] - pc) * t ** x
    plt.plot(xx, yy)  # , markersize=1)
plt.xlabel('$(p - p_c)t^x$')
plt.ylabel('$\\langle r^2 \\rangle t^{-2k}$')
plt.show()
```

Vi ser at det er et data kollaps for $p < p_c$.

```python
beta = 5 / 36
nu = 4 / 3
mu = mu_p2
k = 0.5 * (1 - mu / (2 * nu + mu - beta))
x = (1 - 2 * k) / mu

for i, r2 in enumerate(disps2):
    t = np.arange(len(r2))[1:]
    xx = r2[1:] * t ** (- 2 * k)
    yy = (p2[i] - pc) * t ** x
    plt.plot(yy, xx, label=f'{p2[i]:.2f}')  # 'o', markersize=2)
plt.xlabel('$(p-p_c)t^x$')
plt.ylabel('$r^2t^{-2k}$')
plt.title('$Data collapse for p > p_c$')
plt.show()
```
